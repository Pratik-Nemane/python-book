# 1.Basic Overview of Strings 

<div style="text-align: left; font-size: 1.5em;">
In Python, a string is a data type used to represent a sequence of characters. Characters can include letters, numbers, symbols, and spaces. Strings are immutable, which means once they are created, their content cannot be changed.
</div>

## Characters used in string

<div style="text-align: left; font-size: 1.5em;">
    
`Letters`: The alphabet, a through z and A through Z (lowercase and uppercase)

`Symbols`: All other characters like ! and : and even the space character

`Numbers`: The digits 0 to 9
    


</div>

## Escape Characters
<div style="text-align: left; font-size: 1.5em;">
Escape characters in strings are special sequences that represent characters not easily typed or that have special meanings. They are created using a backslash (\) followed by another character. Here are some common escape characters:

\\ : Backslash

\' : Single quote

\" : Double quote

\n : Newline

\t : Tab

\r : Carriage return

\b : Backspace

\0 : Null character


</div>

## Let's Learn How to Use Escape Characters in Code

````python
# Backslash
print("This is a backslash: \\")

````

````python
# Null character
print("Hello\0World")  # Null character, often used as a string terminator in C/C++
````

````python


# Bell/alert
print("Hello\aWorld")  # Triggers a beep sound in some systems
````

````python
# Backspace
print("Hello\bWorld")  # "HellWorld"
````

````python
# Single quote
print('It\'s a sunny day')
````

````python

# Double quote
print("She said, \"Hello!\"")
````

````python
# Carriage return
print("Hello\rWorld")  # "World" overwrites "Hello"
````

````python
# Tab
print("Column1\tColumn2\tColumn3")
````

````python
# Newline
print("First line\nSecond line")
````

## String Indexing and Slicing

<div style="text-align: left; font-size: 1.5em;">
String indexing and slicing are fundamental operations in Python for accessing and manipulating substrings within a string. Heres an explanation of how indexing and slicing work:
</div>



## String Indexing 

<div style="text-align: left; font-size: 1.5em;">
String indexing allows you to access individual characters within a string using their position, or index. In Python, indexing starts at 0 for the first character of the string.
</div>

                                                                                                                                                         

````python
# Define a string
my_string = "Python"

# Accessing individual characters using positive indexing
first_character = my_string[0]  # 'P'
second_character = my_string[1]  # 'y'
last_character = my_string[-1]  # 'n'

print(first_character, second_character, last_character)
````

## String Slicing

<div style="text-align: left; font-size: 1.5em;">
String slicing in Python allows you to extract a portion of a string based on indices. Heres how it works:
</div>


<div style="text-align: left; font-size: 1.5em;">

#### string[start:end]

</div>





## Let's understand Slicing in strings with the help of few examples.

## Positive Index Slicing

<div style="text-align: left; font-size: 1.2em;">
 In positive indexing, the index starts from 0 for the first character.

For example:

    s[0] returns 'H'
    s[1] returns 'E'
    s[6] returns 'W'

Slicing: Positive slicing allows you to extract a substring from a string. Syntax: s[start:end:step].

    s[0:5] returns "HELLO"
    s[6:] returns "WORLD"
    s[:5] returns "HELLO"


</div>

````python
### postive index slicing

# Define the string
s = "HELLO WORLD"

# Print "hello" using positive indices
print(s[1:4])  
````

````python
### postive index slicing

# Define the string
s = "HELLO WORLD"

# Print "hello" using positive indices
print(s[0:8])  # Output: "hello"
````



## Negative index slicing


<div style="text-align: left; font-size: 1.2em;">

Negative indexing starts from -1 for the last character in the string.


 <div style="text-align: left; font-size: 1.2em;">

In negative indexing, the index starts from -1 for the last character. For example: In negative indexing, the index starts from -1 for the last character. For example:

    s[-1] returns 'D'
    s[-2] returns 'L'
    s[-5] returns 'W'

Slicing: Negative slicing also allows you to extract a substring from a string. Syntax: s[start:end:step].

    s[-5:] returns "WORLD"
    s[:-6] returns "HELLO"
    s[::-1] returns "DLROW OLLEH" (reverses the string)
    
</div>

````python
# Define the string
s = "HELLO WORLD"

# Print "world" using negative indices
print(s[:-6])  # Output: "world"
````

````python
# Define the string
s = "hello world"

# Print "world" using negative indices
print(s[-5:])  # Output: "world"
````

## Slice with Step Size:

### string[start:end:step]
#### The slicing syntax s[start:end:step] is used here.

````python
s = "Hello World!"
print(s[::2])  # Output: "Hlo o!"
````


<div style="text-align: left; font-size: 1.2em;">
Explanation:

    String Definition:
    s = "Hello World!" assigns the string "Hello World!" to the variable s.

    Slicing Expression (s[::2]):
        The slicing syntax s[start:end:step] is used here.
        start is omitted (::), which defaults to the beginning of the string (0).
        end is also omitted (::), which defaults to the end of the string.
        step is 2, indicating that every second character is selected.
</div>

````python
s = "Hello World!"
print(s[::3])  
````

## String Methods


<div style="text-align: left; font-size: 1.5em;">


Python provides a variety of built-in methods for string manipulation. These methods help you perform common operations like formatting, modifying, and querying strings. Here are some of the most commonly used string methods:
</div>

## 1.upper and lower: Convert the string to uppercase or lowercase.

````python
message = "Hello, World!"
print(message.upper())  # Output: HELLO, WORLD!
print(message.lower())  # Output: hello, world!
````

## 2.strip: Remove leading and trailing space.

````python
spaced_string = "  Hello, World!  "
print(spaced_string.strip())  # Output: Hello, World!
````

## 3.replace: Replace parts of the string with another string.

````python
text = "I like apples"
new_text = text.replace("apples", "bananas")
print(new_text)  # Output: I like bananas
````

## 4.Split and Join: Split a string into a list of substrings and join a list of strings into a single string.


<div style="text-align: left; font-size: 1.4em;">

 The split() method splits a string into a list of substrings based on a specified separator. If no separator is provided, the method splits the string at any whitespace (space, tab, newline, etc.). 
</div>

````python
# Example 1: Splitting a string at spaces
s = "Hello World! Python is awesome."
split_list = s.split()
print(split_list)  # Output: ["Hello", "World!", "Python", "is", "awesome."]

# Example 2: Splitting a string with a specified separator
s = "apple,banana,cherry"
split_list = s.split(",")
print(split_list)  # Output: ["apple", "banana", "cherry"]
````


<div style="text-align: left; font-size: 1.4em;">

The join() method joins the elements of an iterable (e.g., a list or tuple) into a single string, with the string on which join() is called acting as the separator.
</div>

````python
# Example 1: Joining a list of strings with a space
words = ["Hello", "World!", "Python", "is", "awesome."]
joined_string = " ".join(words)
print(joined_string)  # Output: "Hello World! Python is awesome."

# Example 2: Joining a list of strings with a comma
fruits = ["apple", "banana", "cherry"]
joined_string = ",".join(fruits)
print(joined_string)  # Output: "apple,banana,cherry"


````

## String Formatting


<div style="text-align: left; font-size: 1.5em;">

String formatting is a way to construct new strings by embedding variables and expressions into a template. Python provides several ways to format strings, each with its own advantages.

#### Syntax

"format_string" % (values)


</div>

````python
name = "Pratik"
age = 29
formatted_string = "Name: %s, Age: %d" % (name, age)
print(formatted_string)  # Output: "Name: Alice, Age: 30"
````
